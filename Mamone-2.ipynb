Lo que haremos a continuación es construir cada índice modificado del sp 500

```python
#Posible punto 1
import warnings
warnings.filterwarnings('ignore')
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
sp_500 = table[0]
tickers = sp_500.Symbol.to_list()

changes = pd.read_excel("sp500change.xlsx")
 

changes["DateAfterChange"] = pd.to_datetime(changes["DateAfterChange"], format="%Y%m%d")
changes.set_index("DateAfterChange", inplace=True)
changes = changes.sort_index(ascending=False)


changes = changes.groupby([pd.Grouper(freq="Q"), "Change"])["Ticker"].agg(lambda x: list(x)).to_frame()

changes = changes.reset_index(drop=False).set_index("DateAfterChange")
changes["Change"] = changes["Change"].map({-1: "drop", 1: "add", 0: "none"})
changes = changes.pivot(columns="Change")
changes.columns = changes.columns.droplevel(0)
tick_series = pd.Series({pd.to_datetime("today"): tickers})
tick_series = tick_series.append(pd.Series(index=changes.index)).sort_index(ascending=False)

for i in tick_series.iloc[1:].index:
    tick_series.loc[i] = list(set(tick_series.shift(1).loc[i] + changes.loc[i]["drop"]).difference(set(changes.loc[i]["add"])))

```
